```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
oscillator = []
movingaverage = []
```

```python
for i in range(3):
    PATH = "C:\\Users\\60167\\Downloads\\chromedriver_win32\\chromedriver.exe"
    driver = webdriver.Chrome(PATH)
    driver.get("https://www.tradingview.com/symbols/EURUSD/technicals/")
    driver.maximize_window()
    search = driver.find_element_by_id("1m")
    driver.execute_script("arguments[0].click();", search)
    html = driver.page_source
    soup = BeautifulSoup(html, "lxml") 
    spanning=soup.find('span', {'class': 'js-symbol-lp-time'})
    oscillatornames = driver.find_elements_by_xpath('//div/div/div[3]/div[1]/div[2]/table/tbody/tr/td[1]')
    oscillatorvalues = driver.find_elements_by_xpath('//div/div/div[3]/div[1]/div[2]/table/tbody/tr/td[2]')
    oscillatoractions = driver.find_elements_by_xpath('//div/div/div[3]/div[1]/div[2]/table/tbody/tr/td[3]')
    mvnames = driver.find_elements_by_xpath('//div/div/div[3]/div[2]/div[2]/table/tbody/tr/td[1]')
    mvvalues = driver.find_elements_by_xpath('//div/div/div[3]/div[2]/div[2]/table/tbody/tr/td[2]')
    mvactions = driver.find_elements_by_xpath('//div/div/div[3]/div[2]/div[2]/table/tbody/tr/td[3]')
    oscillatorsell = driver.find_element_by_xpath ('//div/div/div[2]/div[1]/div[2]/div[1]/span[1]')
    oscillatorneutral = driver.find_element_by_xpath ('//div/div/div[2]/div[1]/div[2]/div[2]/span[1]')
    oscillatorbuy = driver.find_element_by_xpath ('//div/div/div[2]/div[1]/div[2]/div[3]/span[1]')
    mvsell = driver.find_element_by_xpath('//div/div/div[2]/div[3]/div[2]/div[1]/span[1]')
    mvneutral = driver.find_element_by_xpath('//div/div/div[2]/div[3]/div[2]/div[2]/span[1]')
    mvbuy = driver.find_element_by_xpath('//div/div/div[2]/div[3]/div[2]/div[3]/span[1]')
    for i in range(11):
        oscillator_data = {'Time': spanning.text,
                           'SELL': oscillatorsell.text,
                          'NEUTRAL': oscillatorneutral.text,
                          'BUY': oscillatorbuy.text,
                          'Name': oscillatornames[i].text,
                          'Value': oscillatorvalues[i].text,
                          'Action': oscillatoractions[i].text}
        oscillator.append(oscillator_data)
    for i in range(15):
        movingaverage_data = {'Time': spanning.text,
                              'SELL': mvsell.text,
                              'NEUTAL': mvneutral.text,
                             'BUY': mvbuy.text,
                             'Name': mvnames[i].text,
                             'Value': mvvalues[i].text,
                             'Action': mvactions[i].text}
        movingaverage.append(movingaverage_data)
    
    time.sleep(58)
    driver.quit()
```
